### **(a)** Given $L_1$ and $L_2$ are two lower triangular matrices of size $n\times n$, prove that $L_1L_2$ is also a lower triangular matrix. Further, prove by induction that multiplication of $m\,(m>2)$ lower triangular matrices ($L_1, L_2,..., L_m$) is also a lower triangular matrix. (6 points)

#### <font color="red">Your answer here:</font>

Let the $i^{\text{th}}$ row and $j^{\text{th}}$ column of an $n \times n$ matix be:

$$
i, j \in \{1 ... n\} \tag{1}
$$

Definition of a lower triangular matrix:

$$
\mathbf{L_{1_{ij}}} = \mathbf{L_{2_{ij}}} =0\quad\forall\quad i < j \tag{2}
$$

Definition of matrix multiplication:

$$
\mathbf{L_{1} L_{2}} = \mathbf{L_{3_{ij}}} = \sum_{k=1}^{n} \mathbf{L_{1_{ik}} L_{2_{kj}}} \tag{3}
$$

For all $k$, either $\mathbf{L_{1_{ik}}}$ or $\mathbf{L_{2_{kj}}}$ is zero given **(2)** $i<j$: 

$$
k > i \implies \mathbf{L_{1_{ik}}} = 0 \\
k \leq i < j \implies \mathbf{L_{2_{kj}}} = 0 \tag{4}
$$

The matrix product $\mathbf{L_{1} L_{2}}$ **(3)** satisfies the defintion of a lower triangular matrix **(2)**:

$$
\mathbf{L_{3_{ij}}} = \sum_{k=1}^{n} \mathbf{L_{1_{ik}} L_{2_{kj}}} = 0\quad\forall\quad i < j \tag{5}
$$

By induction, the product of any number of $m$ lower triangular matrices ($\mathbf{L_1, L_2,..., L_m}$) is a lower triangular matrix. For $m=3$, (e.g. $\mathbf{L_1, L_2, L_3}$):

                                                                                    
$$
\begin{align}
\mathbf{L_{1_{ij}} L_{2_{ij}} = L_{12_{ij}}} & = 0\quad\forall\quad i < j \\
\mathbf{L_{12_{ij}} L_{3_{ij}} = L_{4_{ij}}} & = 0\quad\forall\quad i < j \\
\mathbf{L_{1_{ij}} L_{2_{ij}} L_{3_{ij}}} & = 0\quad\forall\quad i < j
\end{align}
\tag{6}
$$
